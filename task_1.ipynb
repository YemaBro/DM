```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 读取CSV文件
data = pd.read_csv("../Data/Task_1/data.csv")
data.info()
```

```python
data = data.drop(columns=['Unnamed: 0', 'custid', 'trade_no', 'bank_card_no', 'id_name', 'source', 'student_feature', 'reg_preference_for_trad', 'latest_query_time', 'loans_latest_time'])
```

```python
data.info()
```

```python
data.describe()
```

```python
data.dtypes.value_counts()
```

```python
data = data.dropna(axis=1,thresh=3328)
data = data.dropna(axis=0,thresh=56)
```

```python
data
```

```python
data.isnull().sum(axis=1).sort_values()
```

```python
data = data.fillna(data.mean())
```

```python
data
```

```python
test_data, train_data = train_test_split(data, test_size=0.3, random_state=2018)
```

```python
train_data.to_csv('../Data/Task_1/train_data.csv', index=False, header=True)
```

```python
test_data.to_csv('../Data/Task_1/test_data.csv', index=False, header=True)
```
